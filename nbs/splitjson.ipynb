```python
import json

# Define the path to the JSON file
file_path = './ingest_data_sources/travel-advice-all-countries.json'
data = []
# Load the JSON data from the file
with open(file_path, 'r') as file:
    data = json.load(file)

# Now `data` contains the JSON data and you can work with it.
print(data[0])

one_country = data[0]
```

```python
import json
import os

# Assuming `data[0]` is a list of dictionaries, where each dictionary represents a country
countries_data = data

# Loop through each country's data
for country in countries_data:
    # Extract the country slug to use as the filename
    country_slug = country['Country Slug']
    file_name = f"{country_slug}.json"
    
    # Define the directory where you want to save the files
    output_dir = './countries/country_files'
    os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist
    
    # Define the full path for the new file
    file_path = os.path.join(output_dir, file_name)
    
    # Write the country's data to a new JSON file
    with open(file_path, 'w') as file:
        json.dump(country, file, indent=4)  # Using indent for pretty-printing

print("Files created for each country.")
```

```python
country_index = {}

# Assuming `data` is your list of country dictionaries
for country in data:
    country_name = country.get("Country Name", "")
    base_path = country.get("countryInfo", {}).get("base_path", "")
    
    # Prepend the base_path with the required URL
    full_base_path = f"https://www.gov.uk{base_path}"
    
    # Add the country to the country_index dictionary
    country_index[country_name] = {"base_path": full_base_path}

# To use the country_index or to write it to a file:
print(country_index)

# If you need to write this to a JSON file:
with open('country_index.json', 'w') as file:
    json.dump(country_index, file, indent=4)
```
